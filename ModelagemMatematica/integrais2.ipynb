```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

import mpl_toolkits.mplot3d.axes3d as axes3d
```

```python
x = sym.Symbol('x')
sym.init_printing(pretty_print=True)
g0 = x**3
g1 = 1
g2 = 8
f = 0
```

```python
plot = sym.plot(g0, g1, g2, f, legend = True)
```

```python
def volume_solido_disco(funcao, x_i, x_f):
    integral_def = sym.integrate((funcao)**2, (x, x_i, x_f))
    volume = np.pi*integral_def
    return volume.round(3)
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(1, 1, 1, projection='3d')

u = np.linspace(-2, 1, 60)
v = np.linspace(0, 2*np.pi, 60)
U, V = np.meshgrid(u, v)

X = U
Y1 = (U**2 + 1)*np.cos(V)
Z1 = (U**2 + 1)*np.sin(V)

Y2 = (3 - U)*np.cos(V)
Z2 = (3 - U)*np.sin(V)

ax.plot_surface(X, Y1, Z1, alpha=0.3, color='red', rstride=6, cstride=12)
ax.plot_surface(X, Y2, Z2, alpha=0.3, color='blue', rstride=6, cstride=12)
plt.show()
```
