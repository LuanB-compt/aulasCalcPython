```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as axes3d

import gif
from IPython.display import Image
```

# EXERC√çCIO 1

```python
x = sym.Symbol('x')
f = (x**2)/2

def funcao(input):
    return (input**2)/2

print(funcao(3.16228))
```

```python
sym.plot(f, (x, 0, 3.16228))
```

```python
y = sym.Symbol('y')
sym.plot_implicit((y > f), (x, 0, 3.16228), title='EXERC√çCIO 1') # podemos fazer algumas opera√ß√µes l√≥gicas
```

```python
sym.plot(((x**2 + 1)), (x, -2, 2), ylim=[-5, 10])
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')

u = np.linspace(0, 3.16228, 50)
v = np.linspace(0, 2*np.pi, 50)
U, V = np.meshgrid(u, v)

X = U
Y = ((U**2)/2)*np.cos(V)
Z = ((U**2)/2)*np.sin(V)

ax.plot_surface(X, Y, Z, alpha=0.3, color='red', rstride=6, cstride=12)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
