```python
import sympy as sp
from sympy import *
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# EXERC√çCIO 1

```python
sp.init_printing(pretty_print=True)
x, y, z, t = sp.Symbol('x'), sp.Symbol('y'), sp.Symbol('z'), sp.Symbol('t')
#f, g, h = map(Function, 'fgh')
f = (x**2) - 4
g = ((x**2)*(-1)) - (2*x)
```

```python
f_plot = sp.plot(f, xlim=[-4, 4], ylim=[-5, 2])
g_plot = sp.plot(g, xlim=[-4, 4], ylim=[-5, 2])

grafico_junto = f_plot
grafico_junto.append(g_plot[0]) #sp.plot(f, g, xlim=[-4, 4], ylim=[-5, 2])
grafico_junto.show()
```

```python
intervalo = sp.solve(f - g)
print(intervalo)
print(type(intervalo[0]))
```

```python
print(type(float(intervalo[0])))
print(float(intervalo[0]))
```

```python
integral_definida = sp.integrate(f - g, (x, -2, 1))
integral_definida
```

# EXERC√çCIO 2

```python
x = sp.Symbol('x')
f = x**(1/3)
g = x**(1/2)
```

```python
f_plot = sp.plot(f, (x, 0, 2))
g_plot = sp.plot(g, (x, 0, 2))

grafico_junto = f_plot
grafico_junto.append(g_plot[0])
grafico_junto.show()
```

# EXERC√çCIO 3

```python
x = sp.Symbol('x')
f = x**2
g = 2-x
h = 0
```

```python
intervalo = sp.solve(f - g - h)
intervalo
```

```python
f_plot = sp.plot(f, (x, -3, 3))
g_plot = sp.plot(g, (x, -3, 3))
h_plot = sp.plot(h, (x, -3, 3))

grafico_junto = f_plot
grafico_junto.append(g_plot[0])
grafico_junto.append(h_plot[0])
grafico_junto.legend = True
grafico_junto.show()
```

```python
integral_definida = sp.integrate(f-h, (x, 0, 1))
abs(float(integral_definida))
```

```python
integral_definida_2 = sp.integrate(g-h, (x, 1, 2))
abs(float(integral_definida_2))
```

```python
integral_definida_final = abs(float(integral_definida)) + abs(float(integral_definida_2))
float(integral_definida_final)
```
