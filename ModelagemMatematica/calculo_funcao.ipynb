### **FUNÇÃO QUE PLOT GRÁFICO DE QUALQUER FUNÇÃO COM MATPLOTLIB:**

```python
def equacao(input):
    output = (2*pi*(input**2)) + (2000/input)
    return output

def desenho_funcao(range_x):
    x = []
    for i in range_x:
        if i != 0:
            x.append(i)

    y = []
    for i in x:
        y.append(equacao(i))

    plt.figure(figsize=(10,7))
    plt.scatter(x, y)
    plt.show()

desenho_funcao(range(-8, 12))
```

#### - Substituindo a altura pela relação matemática:
- desse modo temos apenas uma variável nas funções, facilitando a resolução

```python
altura = 1000/(pi*raio**2)
raio = symbols("r")

perimetro_circunferencia = 2*pi*raio
volume_cilindro = (pi*(raio**2))*altura
area_cilindro = pi*raio**2
area_corpo_cilindro = perimetro_circunferencia*altura
area_total = 2*area_cilindro + area_corpo_cilindro

area_total
```

```python
area_total_dr = diff(area_total)
area_total_dr
```

```python
area_total_derivada_seg = diff(area_total_dr)
area_total_derivada_seg
```

#### - Plotando os gráficos de cada função:
- análise visual de cada um separado
- análise visual de todos eles juntos

```python
grafico_area_total = plot(area_total, xlim=[-8, 5], ylim=[-2000, 2000], legend=True, size=(10,5))
grafico_area_total_dr = plot(area_total_dr, xlim=[-8, 5], ylim=[-5000, 1], legend=True, size=(10,5))
grafico_area_total_derivada_seg = plot(area_total_derivada_seg, xlim=[-8, 5], ylim=[-2000, 2000], legend=True, size=(10,5))
```

```python
graficos_juntos = grafico_area_total
graficos_juntos.append(grafico_area_total_dr[0])
graficos_juntos.append(grafico_area_total_derivada_seg[0])

graficos_juntos.title = 'Todos gráficos juntos\n'
grafico_area_total.size = (12,7)
graficos_juntos.show()
```
